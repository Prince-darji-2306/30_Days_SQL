#### 1. Total Sales by Region
Write a query to find the total sales for each region.

````python
cursor.execute("DROP TABLE IF EXISTS sales")
cursor.execute('''
CREATE TABLE sales (
    sale_id INT PRIMARY KEY,
    salesperson VARCHAR(50),
    region VARCHAR(50),
    sales_amount DECIMAL(10,2),
    sale_date DATE
)
''')

sales_data = [
    (1, 'Alice', 'North', 5000, '2024-01-01'),
    (2, 'Bob', 'South', 7000, '2024-01-02'),
    (3, 'Alice', 'North', 3000, '2024-01-03'),
    (4, 'Bob', 'South', 2000, '2024-01-04'),
    (5, 'Carol', 'East', 8000, '2024-01-05')
]
cursor.executemany("INSERT INTO sales VALUES (%s, %s, %s, %s, %s)", sales_data)
conn.commit()

query = """
SELECT region, SUM(sales_amount) AS total_sales
FROM sales
GROUP BY region;
"""
pd.read_sql(query, conn)
````

#### 2. Average Score by Class
Write a query to find the average score for each class.

````python
cursor.execute("DROP TABLE IF EXISTS students")
cursor.execute('''
CREATE TABLE students (
    student_id INT PRIMARY KEY,
    student_name VARCHAR(50) NOT NULL,
    class VARCHAR(10),
    course VARCHAR(50),
    score INT
)
''')

student_data = [
    (1, 'John', 'A', 'Math', 85),
    (2, 'Mary', 'A', 'Math', 90),
    (3, 'Jake', 'B', 'Science', 70),
    (4, 'Emily', 'B', 'Science', 75),
    (5, 'Steve', 'A', 'Math', 95)
]
cursor.executemany("INSERT INTO students VALUES (%s, %s, %s, %s, %s)", student_data)
conn.commit()

query = """
SELECT class, AVG(score) AS average_score
FROM students
GROUP BY class;
"""
pd.read_sql(query, conn)
````

#### 3. High-Selling Salespersons
Write a query to find salespersons with total sales greater than 8000.

````python
query = """
SELECT salesperson, SUM(sales_amount) AS total_sales
FROM sales
GROUP BY salesperson
HAVING total_sales > 8000;
"""
pd.read_sql(query, conn)
````

#### 4. Students with More Than One Entry
Write a query to find all students who appear more than once in the table.

````python
query = """
SELECT student_name, COUNT(*) AS entry_count
FROM students
GROUP BY student_name
HAVING entry_count > 1;
"""
pd.read_sql(query, conn)
````
