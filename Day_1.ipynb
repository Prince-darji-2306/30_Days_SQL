#### 1. Database Connection
We will connect to MySQL and ensure the 30_Days_SQL database exists.

```python
import mysql.connector

try:
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password=""
    )
    cursor = conn.cursor()

    cursor.execute("CREATE DATABASE IF NOT EXISTS 30_Days_SQL")
    cursor.execute("USE 30_Days_SQL")
    
    print("Database '30_Days_SQL' is ready!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
```

#### 2. Task for Day 1: Run Your First SQL Query
Execute a simple SELECT statement to confirm everything is working.

```python
cursor.execute("SELECT 'Hello, SQL!' AS Message")
result = cursor.fetchone()
print(result[0])
```
