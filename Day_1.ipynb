### 1. Database Connection
We will connect to MySQL and ensure the `sql_30days_challenge` database exists.

```python
import mysql.connector

try:
    # Connect to MySQL (initial connection without database to create it if needed)
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password=""  # Default XAMPP/MySQL password
    )
    cursor = conn.cursor()

    # Create database if not exists
    cursor.execute("CREATE DATABASE IF NOT EXISTS sql_30days_challenge")
    cursor.execute("USE sql_30days_challenge")
    
    print("Database 'sql_30days_challenge' is ready!")
except mysql.connector.Error as err:
    print(f"Error: {err}")
```

### 2. Task for Day 1: Run Your First SQL Query
Execute a simple `SELECT` statement to confirm everything is working.

```python
cursor.execute("SELECT 'Hello, SQL!' AS Message")
result = cursor.fetchone()
print(result[0])
```
