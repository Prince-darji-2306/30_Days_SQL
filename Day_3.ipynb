#### Day 3: Filtering Data with WHERE Clause 🌟

Today we focus on filtering records using the WHERE clause.

#### 1. Database Connection

```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="30_Days_SQL"
)
cursor = conn.cursor()
print("Connected to '30_Days_SQL'!")
```

---

#### Practice Exercises 📘

#### 1. Retrieve all employees whose age is less than 30

```python
query = "SELECT * FROM employees WHERE age < 30"
pd.read_sql(query, conn)
```

#### 2. Find employees in the Marketing department or aged 35 and above

```python
query = "SELECT * FROM employees WHERE department = 'Marketing' OR age >= 35"
pd.read_sql(query, conn)
```

#### 3. Retrieve employees whose names end with the letter 'n'

```python
query = "SELECT * FROM employees WHERE name LIKE '%n'"
pd.read_sql(query, conn)
```

#### 4. List employees who do not belong to the IT department

```python
query = "SELECT * FROM employees WHERE department != 'IT'"
pd.read_sql(query, conn)
```

#### 5. Retrieve employees with non-NULL age values

```python
query = "SELECT * FROM employees WHERE age IS NOT NULL"
pd.read_sql(query, conn)
```
