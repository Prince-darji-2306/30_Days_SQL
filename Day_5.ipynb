### 1. Total number of students in each course
(We'll add a `course` column to students and update data)

```python
try:
    cursor.execute("ALTER TABLE students ADD COLUMN course VARCHAR(50)")
except:
    pass

cursor.execute("UPDATE students SET course = 'SQL' WHERE student_id = 101")
cursor.execute("UPDATE students SET course = 'Python' WHERE student_id = 102")
cursor.execute("UPDATE students SET course = 'SQL' WHERE student_id = 103")
conn.commit()

query = "SELECT course, COUNT(*) AS student_count FROM students GROUP BY course"
pd.read_sql(query, conn)
```

### 2. Average salary of employees in each department

```python
query = "SELECT department, AVG(salary) AS average_salary FROM employees GROUP BY department"
pd.read_sql(query, conn)
```

### 3. Minimum and maximum price of products in each category
(We'll add a `category` column to products)

```python
try:
    cursor.execute("ALTER TABLE products ADD COLUMN category VARCHAR(50)")
except:
    pass

cursor.execute("UPDATE products SET category = 'Electronics' WHERE product_id IN (1, 4)")
cursor.execute("UPDATE products SET category = 'Accessories' WHERE product_id IN (2, 3, 5)")
conn.commit()

query = "SELECT category, MIN(price) AS min_price, MAX(price) AS max_price FROM products GROUP BY category"
pd.read_sql(query, conn)
```

### 4. Count the number of orders placed by each customer
(First, create the `orders` table)

```python
cursor.execute("DROP TABLE IF EXISTS orders")
cursor.execute('''
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    customer_id INT,
    order_date DATE,
    order_total DECIMAL(10, 2)
)
''')

orders_data = [
    (1, 101, '2023-10-01', 150.00),
    (2, 102, '2023-10-02', 200.00),
    (3, 101, '2023-10-03', 50.00),
    (4, 103, '2023-10-04', 300.00),
    (5, 102, '2023-10-05', 100.00)
]
cursor.executemany("INSERT INTO orders VALUES (%s, %s, %s, %s)", orders_data)
conn.commit()

query = "SELECT customer_id, COUNT(*) AS order_count FROM orders GROUP BY customer_id"
pd.read_sql(query, conn)
```
