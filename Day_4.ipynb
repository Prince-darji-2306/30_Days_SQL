### 1. Database Connection

```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="sql_30days_challenge"
)
cursor = conn.cursor()
print("Connected to 'sql_30days_challenge'!")
```

## Practice Exercises ðŸ–‹

### 1. Retrieve all employees and sort them by `name` in ascending order

```python
query = "SELECT * FROM employees ORDER BY name ASC"
pd.read_sql(query, conn)
```

### 2. Sort the `employees` table by `department` in descending order

```python
query = "SELECT * FROM employees ORDER BY department DESC"
pd.read_sql(query, conn)
```

### 3. Combine sorting by `age` in ascending order and `department` in alphabetical order

```python
query = "SELECT * FROM employees ORDER BY age ASC, department ASC"
pd.read_sql(query, conn)
```

### 4. Sort `products` by `price` in descending and `stock_quantity` in ascending order
(First, let's insert data into products)

```python
products_data = [
    (1, 'Laptop', 1200.00, 10),
    (2, 'Mouse', 25.50, 50),
    (3, 'Keyboard', 75.00, 20),
    (4, 'Monitor', 200.00, 15),
    (5, 'USB Cable', 10.00, 100)
]
cursor.executemany("INSERT IGNORE INTO products VALUES (%s, %s, %s, %s)", products_data)
conn.commit()

query = "SELECT * FROM products ORDER BY price DESC, stock_quantity ASC"
pd.read_sql(query, conn)
```
