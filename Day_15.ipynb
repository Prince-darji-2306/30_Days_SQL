#### Practice Exercises 🛠️

#### 1. String Aggregation (GROUP_CONCAT)
Create a comma-separated list of product names for each category.

```python
query = """
SELECT category, GROUP_CONCAT(product_name SEPARATOR ', ') AS product_list
FROM products
GROUP BY category
"""
pd.read_sql(query, conn)
```

#### 2. Statistical Functions (STDDEV and VARIANCE)
Calculate the standard deviation and variance of employee salaries.

```python
query = """
SELECT 
    department_id, 
    STDDEV(salary) AS salary_stddev, 
    VARIANCE(salary) AS salary_variance
FROM employees
GROUP BY department_id
"""
pd.read_sql(query, conn)
```

#### 3. Hierarchical Aggregation (ROLLUP)
Calculate subtotals and grand totals for sales by region.

```python
query = """
SELECT region, SUM(sales_amount) AS total_sales
FROM sales
GROUP BY region WITH ROLLUP
"""
pd.read_sql(query, conn)
```
